```python
folder_path = "C:\\Users\\Guilherme Lube\\Documents\\Trabalho\\Download_licitacao\\licitacoes_compras\\licitacoes3"
for (current_folder, folder_in_current_folder, files_in_current_folder) in os.walk(folder_path):
    for filename in files_in_current_folder:
        print(filename)
        print(current_folder)
        if filename.endswith('.pdf'):
            if filename != 'Edital 092.2024 - Iluminação Pública - 1ª Alteração ASSINADO.pdf' and filename != 'PRJ-EXE-EST-CAM-PPV-REV00-0308.pdf' and filename != 'PRJ-EXE-EST-CAM-PPV-REV00-0408.pdf':
                pdf_file = open(os.path.join(current_folder, filename), 'rb')
                pdf_reader = PyPDF2.PdfReader(pdf_file)
                text = ''
                for i in range(len(pdf_reader.pages)):
                        #text = ''
                    page = pdf_reader.pages[i]
                    text += page.extract_text()
                    txt_filename = os.path.splitext(filename)[0] + '.txt'
                    txt_file = open(os.path.join(current_folder, txt_filename), 'w', encoding="utf-8")
                    txt_file.write(text)
                i = 0
                pdf_file.close()
                txt_file.close()
```
